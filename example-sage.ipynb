## Задача № 1
Найдите решение системы уравнений
$$
x^2+y^2+z=1, \quad x^2-y^2+z=x, \quad x^2=y+z
$$

Решение. Задаем неизвестные

```python
var("x,y,z")
```

Задаем систему

```python
eqs=[x^2+y^2+z-1, x^2-y^2+z-x, x^2-y-z]
```

Решение

```python
K=QQ[x,y,z]
J=K*eqs
```

```python
J
```

```python
J.groebner_basis()
```

```python
J.variety(QQbar)
```
